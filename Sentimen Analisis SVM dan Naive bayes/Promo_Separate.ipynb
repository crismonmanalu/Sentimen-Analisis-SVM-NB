```python
#Import libraries
import numpy as np 
import pandas as pd 
import nltk
from nltk.corpus import stopwords
import string
import pickle
```

```python
df = pd.read_csv('datatrain.csv') #read the CSV file
data_label = df["full_text"]
label = df["promo"]
df.head(5)
```

```python
df.drop_duplicates(inplace = True)
```

```python
#Split the data into 80% training (X_train & y_train) and 20% testing (X_test & y_test) data sets
from sklearn.cross_validation  import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_label, label, test_size = 0.20, random_state = 0)
```

```python
# Get the shape of messages_bow
# messages_bow.shape
from sklearn.feature_extraction.text import TfidfVectorizer

# vectorize email text into tfidf matrix
# TfidfVectorizer converts collection of raw documents to a matrix of TF-IDF features.
# It's equivalent to CountVectorizer followed by TfidfTransformer.
vectorizer = TfidfVectorizer(
    input='content',     # input is actual text
    lowercase=True,      # convert to lower case before tokenizing
    stop_words='english' # remove stop words
)
features_train_transformed = vectorizer.fit_transform(X_train)
features_test_transformed  = vectorizer.transform(X_test)
```

```python
#Create and train the Naive Bayes classifier
#The multinomial Naive Bayes classifier is suitable for classification with discrete features (e.g., word counts for text classification)
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(features_train_transformed, y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(features_test_transformed, y_test)
...
# score the classifier accuracy
print("classifier accuracy {:.2f}%".format(classifier.score(features_test_transformed, y_test) * 100))
```

```python
with open('model_promo.pkl', 'wb') as file:
    pickle.dump((vectorizer, classifier), file)
```

```python
def load():
    '''
    load classifier from disk
    '''
    with open('model_promo.pkl', 'rb') as file:
        vectorizer, clf = pickle.load(file)
    return vectorizer, clf
```

```python
vectorizer, classifer = load()
print('\nPerform a test')      
df = pd.read_csv('datafix.csv') #read the CSV file
data = df['full_text']
data_label = df["full_text"]
df.drop_duplicates(inplace = True)
```

```python
tweet_input_transformed = vectorizer.transform(data_label)
prediction = classifer.predict(tweet_input_transformed)
```

```python
import collections, numpy
print("Jumlah Tweet: {}".format(len(data_label)))
collections.Counter(prediction)
```

```python
results = pd.DataFrame({
        "tweet" : data,
        "Data" : data_label,
        "Promo" : prediction
        })
results.to_csv("datafix_promo.csv", index = False)
```
